# Regex 


- * means zero or more
- . means match any character

- \( tell regex to include ( in searching expression

```python
#### "r\(.*\)"
# \( - tell regex to include ( in searching expression
```

```python
sample={"names":['Abhay','Rajeev','Adithya'],
"Occupation":['Stud1','Stud2','Stud3'],
"City":['Bangalore','Mumbai','Delhi']
}
```

```python
sample_df=pd.DataFrame(sample)
```

```python
sample_df
```

```python
sample_df.replace('A.*','CAT',regex=True)
```
